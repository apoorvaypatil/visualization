```python
plt.figure(figsize=(12, 6))
data = pd.read_csv('2015_16_Districtwise.csv')  # Replace 'your_data.csv' with the path to your actual CSV file

ax = sns.barplot(data=data, x='STATNAME', y='TOTPOPULAT')
plt.xlabel('States')
plt.ylabel('Total Population')
plt.title('Population per State in India')
plt.xticks(rotation=90)
# Annotate the exact y-values on the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```
